```python
import cv2
from skimage.feature import graycomatrix, graycoprops, local_binary_pattern
from torch.utils.data import Dataset, DataLoader
from PIL import Image
import numpy as np
```

```python
def extract_glcm_features(gray_img):
    """Extract GLCM texture features from a grayscale image (uint8)."""
    gray_img = (gray_img // 4).astype(np.uint8)  # 64 levels
    distances = [1, 3]
    angles = [0, np.pi/4, np.pi/2, 3*np.pi/4]
    glcm = graycomatrix(gray_img, distances=distances, angles=angles,
                        levels=64, symmetric=True, normed=True)
    props = ['contrast', 'dissimilarity', 'homogeneity', 'energy', 'correlation', 'ASM']
    feats = []
    for prop in props:
        vals = graycoprops(glcm, prop)
        feats.append(vals.mean())
        feats.append(vals.std())
    return np.array(feats, dtype=np.float32)  # 12 features


def extract_lbp_features(gray_img, radius=3, n_points=24, n_bins=26):
    """Extract LBP histogram features from a grayscale image."""
    lbp = local_binary_pattern(gray_img, n_points, radius, method='uniform')
    hist, _ = np.histogram(lbp.ravel(), bins=n_bins, range=(0, n_points + 2), density=True)
    return hist.astype(np.float32)  # 26 features


def extract_radiomics(image_pil):
    """Given a PIL image, return a 1-D numpy array of radiomic features."""
    img_np = np.array(image_pil.convert('RGB'))
    gray = cv2.cvtColor(img_np, cv2.COLOR_RGB2GRAY)
    gray = cv2.resize(gray, (256, 256))
    glcm_feats = extract_glcm_features(gray)
    lbp_feats  = extract_lbp_features(gray)
    return np.concatenate([glcm_feats, lbp_feats])  # 38 total

# Sanity check
_dummy = Image.new('RGB', (256, 256))
_r = extract_radiomics(_dummy)
print(f"Radiomic feature vector length: {len(_r)}")  # expect 38
```

```python
class RadiomicsImageFolder(Dataset):
    """Wraps an ImageFolder dataset and augments each sample with radiomic features."""

    def __init__(self, image_folder_dataset):
        self.dataset = image_folder_dataset

    def __len__(self):
        return len(self.dataset)

    def get_labels(self):
        """Required by ImbalancedDatasetSampler."""
        return [s[1] for s in self.dataset.samples]

    def __getitem__(self, idx):
        img_tensor, label = self.dataset[idx]
        path, _ = self.dataset.samples[idx]
        pil_img = Image.open(path).convert('RGB')
        radiomics = extract_radiomics(pil_img)
        return img_tensor, torch.tensor(radiomics, dtype=torch.float32), label


train_radio_dataset = RadiomicsImageFolder(train_dataset)
val_radio_dataset   = RadiomicsImageFolder(val_dataset)
test_radio_dataset  = RadiomicsImageFolder(test_dataset)

train_radio_loader = DataLoader(train_radio_dataset,
                                sampler=ImbalancedDatasetSampler(train_radio_dataset),
                                batch_size=32)
val_radio_loader   = DataLoader(val_radio_dataset,  batch_size=32, shuffle=False)
test_radio_loader  = DataLoader(test_radio_dataset, batch_size=32, shuffle=False)

print(f"Train: {len(train_radio_dataset)}, Val: {len(val_radio_dataset)}, Test: {len(test_radio_dataset)}")
```

```python
class ImbalancedDatasetSampler(WeightedRandomSampler):
    """Samples elements randomly from a given list of indices with a probability
    proportional to the inverse of the class frequency.

    Args:
        dataset: Dataset for sampling.
        indices: a list of indices
        num_samples: number of samples to draw
        callback_get_label (callable, optional): a callback-like function that takes
            a dataset and an index, and returns the label for the given index
    """
    def __init__(self, dataset, indices=None, num_samples=None, callback_get_label=None):
        # if indices is not provided, all elements in the dataset will be considered
        self.indices = list(range(len(dataset))) if indices is None else indices

        # if num_samples is not provided, draw as many samples as specified in indices
        self.num_samples = len(self.indices) if num_samples is None else num_samples

        # if callback is not provided, use default labels_getter function
        if callback_get_label is None:
            callback_get_label = self._get_labels

        # get labels from the dataset
        labels = callback_get_label(dataset, self.indices)
        unique_labels, counts = np.unique(labels, return_counts=True)
        label_to_count = {label: count for label, count in zip(unique_labels, counts)}

        weights = [1.0 / label_to_count[label] for label in labels]
        super().__init__(weights, self.num_samples, replacement=True)

    def _get_labels(self, dataset, indices):
        return dataset.get_labels()
```

```python
def calculate_accuracy(y_pred, y_true):
    top_pred = y_pred.argmax(1, keepdim=True)
    correct = top_pred.eq(y_true.view_as(top_pred)).sum()
    acc = correct.float() / y_true.shape[0]
    return acc

def performance_matrix(y_true, y_pred):
    cm = confusion_matrix(y_true, y_pred)
    print('Confusion Matrix:')
    print(cm)
    print('Classification Report:')
    print(classification_report(y_true, y_pred))
```

```python
# Placeholder for CBAM, densenet169, checkpoint_model, epoch_time. These might be defined elsewhere or need to be implemented.
# For now, I will provide a minimal definition for CBAM and checkpoint_model to allow the code to run.

class CBAM(nn.Module):
    def __init__(self, gate_channels, reduction_ratio=16, pool_types=['avg', 'max']):
        super(CBAM, self).__init__()
        self.ChannelAttention = ChannelAttention(gate_channels, reduction_ratio, pool_types)
        self.SpatialAttention = SpatialAttention()

    def forward(self, x):
        x_out = self.ChannelAttention(x) * x
        x_out = self.SpatialAttention(x_out) * x_out
        return x_out

class ChannelAttention(nn.Module):
    def __init__(self, gate_channels, reduction_ratio=16, pool_types=['avg', 'max']):
        super(ChannelAttention, self).__init__()
        self.gate_channels = gate_channels
        self.mlp = nn.Sequential(
            Flatten(),
            nn.Linear(gate_channels, gate_channels // reduction_ratio),
            nn.ReLU(),
            nn.Linear(gate_channels // reduction_ratio, gate_channels)
            )
        self.pool_types = pool_types

    def forward(self, x):
        channel_att_sum = None
        for pool_type in self.pool_types:
            if pool_type=='avg':
                avg_pool = F.avg_pool2d( x, (x.size(2), x.size(3)), stride=(x.size(2), x.size(3)))
                channel_att_raw = self.mlp( avg_pool )
            elif pool_type=='max':
                max_pool = F.max_pool2d( x, (x.size(2), x.size(3)), stride=(x.size(2), x.size(3)))
                channel_att_raw = self.mlp( max_pool )
            elif pool_type=='lp':
                lp_pool = F.lp_pool2d( x, 2, (x.size(2), x.size(3)), stride=(x.size(2), x.size(3)))
                channel_att_raw = self.mlp( lp_pool )
            elif pool_type=='lse':
                # LSE pool only
                lse_pool = logsumexp_2d(x)
                channel_att_raw = self.mlp( lse_pool )

            if channel_att_sum is None:
                channel_att_sum = channel_att_raw
            else:
                channel_att_sum = channel_att_sum + channel_att_raw

        scale = torch.sigmoid( channel_att_sum ).unsqueeze(2).unsqueeze(3).expand_as(x)
        return scale

class SpatialAttention(nn.Module):
    def __init__(self, kernel_size=7):
        super(SpatialAttention, self).__init__()
        self.compress = ChannelPool()
        self.spatial = BasicConv(2, 1, kernel_size, stride=1, padding=(kernel_size-1) // 2, relu=False)

    def forward(self, x):
        x_compress = self.compress(x)
        x_out = self.spatial(x_compress)
        scale = torch.sigmoid(x_out).expand_as(x)
        return scale

class Flatten(nn.Module):
    def forward(self, x):
        return x.view(x.size(0), -1)

class ChannelPool(nn.Module):
    def forward(self, x):
        return torch.cat( (torch.max(x,1)[0].unsqueeze(1), torch.mean(x,1).unsqueeze(1)), dim=1 )

class BasicConv(nn.Module):
    def __init__(self, in_planes, out_planes, kernel_size, stride=1, padding=0, dilation=1, groups=1, relu=True, bn=True, bias=False):
        super(BasicConv, self).__init__()
        self.out_channels = out_planes
        self.conv = nn.Conv2d(in_planes, out_planes, kernel_size=kernel_size, stride=stride, padding=padding, dilation=dilation, groups=groups, bias=bias)
        self.bn = nn.BatchNorm2d(out_planes,eps=1e-5, momentum=0.01, affine=True) if bn else None
        self.relu = nn.ReLU() if relu else None

    def forward(self, x):
        x = self.conv(x)
        if self.bn is not None:
            x = self.bn(x)
        if self.relu is not None:
            x = self.relu(x)
        return x

def checkpoint_model(epoch, model, optimizer, best_acc, path):
    # This is a placeholder function for saving model checkpoints
    torch.save({
        'epoch': epoch,
        'model_state_dict': model.state_dict(),
        'optimizer_state_dict': optimizer.state_dict(),
        'best_acc': best_acc,
    }, path)

def epoch_time(start_time, end_time):
    elapsed_time = end_time - start_time
    elapsed_mins = int(elapsed_time / 60)
    elapsed_secs = int(elapsed_time - (elapsed_mins * 60))
    return elapsed_mins, elapsed_secs
```

```python
RADIOMICS_DIM = 38  # 12 GLCM + 26 LBP

class DenseNet169_CBAM_RadiomicsFusion(nn.Module):
    def __init__(self, num_classes=3, radiomics_dim=RADIOMICS_DIM):
        super(DenseNet169_CBAM_RadiomicsFusion, self).__init__()

        # DenseNet169 backbone (pretrained)
        self.densenet = densenet169(pretrained=True)
        self.cnn_feat_dim = self.densenet.classifier.in_features  # 1664
        self.densenet.classifier = nn.Identity()

        # CBAM after initial conv layers (64-ch feature maps)
        self.cbam = CBAM(64)

        # Radiomics branch: small MLP to project radiomics to a learned space
        self.radiomics_branch = nn.Sequential(
            nn.Linear(radiomics_dim, 128),
            nn.BatchNorm1d(128),
            nn.ReLU(inplace=True),
            nn.Dropout(0.3),
            nn.Linear(128, 128),
            nn.BatchNorm1d(128),
            nn.ReLU(inplace=True),
        )

        # Fusion classifier: CNN features (1664) + Radiomics (128) -> classes
        fused_dim = self.cnn_feat_dim + 128
        self.classifier = nn.Sequential(
            nn.Linear(fused_dim, 512),
            nn.BatchNorm1d(512),
            nn.ReLU(inplace=True),
            nn.Dropout(0.4),
            nn.Linear(512, 128),
            nn.BatchNorm1d(128),
            nn.ReLU(inplace=True),
            nn.Dropout(0.3),
            nn.Linear(128, num_classes),
        )

    def forward_cnn(self, x):
        """Run DenseNet+CBAM backbone, return feature maps before global pool."""
        x = self.densenet.features.conv0(x)
        x = self.densenet.features.norm0(x)
        x = self.densenet.features.relu0(x)
        x = self.densenet.features.pool0(x)
        x = self.cbam(x)
        x = self.densenet.features.denseblock1(x)
        x = self.densenet.features.transition1(x)
        x = self.densenet.features.denseblock2(x)
        x = self.densenet.features.transition2(x)
        x = self.densenet.features.denseblock3(x)
        x = self.densenet.features.transition3(x)
        x = self.densenet.features.denseblock4(x)
        x = self.densenet.features.norm5(x)
        x = torch.relu(x)
        return x  # (B, 1664, H, W)

    def forward(self, img, radiomics):
        feat_maps = self.forward_cnn(img)
        cnn_feats = F.adaptive_avg_pool2d(feat_maps, 1).flatten(1)  # (B, 1664)
        radio_feats = self.radiomics_branch(radiomics)               # (B, 128)
        fused = torch.cat([cnn_feats, radio_feats], dim=1)           # (B, 1792)
        out = self.classifier(fused)
        return out


model2 = DenseNet169_CBAM_RadiomicsFusion(num_classes=3).to(device)
print(f"Total trainable parameters: {sum(p.numel() for p in model2.parameters() if p.requires_grad):,}")
```

```python
criterion2 = nn.CrossEntropyLoss()
optimizer2 = optim.Adam(model2.parameters(), lr=1e-3)
scaler2 = torch.amp.GradScaler('cuda', enabled=True)

train_accu2 = []
training_loss2 = []

def train_fusion(model, iterator, optimizer, criterion, device):
    epoch_loss = 0
    epoch_acc = 0
    image_preds_all = []
    image_targets_all = []
    model.train()

    for (imgs, radios, labels) in tqdm(iterator, desc="Training Fusion", leave=False):
        imgs   = imgs.to(device).float()
        radios = radios.to(device).float()
        labels = labels.to(device).long()

        with torch.autocast(device_type='cuda'):
            preds = model(imgs, radios)
            loss  = criterion(preds, labels)

        image_preds_all  += [torch.argmax(preds, 1).detach().cpu().numpy()]
        image_targets_all += [labels.detach().cpu().numpy()]
        acc = calculate_accuracy(preds, labels)

        scaler2.scale(loss).backward()
        scaler2.step(optimizer)
        scaler2.update()
        optimizer.zero_grad()

        epoch_loss += loss.item()
        epoch_acc  += acc.item()

    image_preds_all  = np.concatenate(image_preds_all)
    image_targets_all = np.concatenate(image_targets_all)
    score = (image_preds_all == image_targets_all).mean()

    train_losss = epoch_loss / len(iterator)
    train_accu2.append(score * 100)
    training_loss2.append(train_losss)

    return epoch_loss / len(iterator), epoch_acc / len(iterator), train_accu2, training_loss2


val_accu2 = []
eval_loss2 = []

def evaluate_fusion(model, iterator, criterion, device):
    epoch_loss = 0
    epoch_acc = 0
    image_preds_all = []
    image_targets_all = []
    model.eval()

    with torch.no_grad():
        for (imgs, radios, labels) in tqdm(iterator, desc="Evaluating Fusion", leave=False):
            imgs   = imgs.to(device).float()
            radios = radios.to(device).float()
            labels = labels.to(device).long()

            preds = model(imgs, radios)
            loss  = criterion(preds, labels)
            acc   = calculate_accuracy(preds, labels)

            image_preds_all  += [torch.argmax(preds, 1).detach().cpu().numpy()]
            image_targets_all += [labels.detach().cpu().numpy()]
            epoch_loss += loss.item()
            epoch_acc  += acc.item()

    image_preds_all  = np.concatenate(image_preds_all)
    image_targets_all = np.concatenate(image_targets_all)
    score = (image_preds_all == image_targets_all).mean()

    val_losss = epoch_loss / len(iterator)
    val_accu2.append(score * 100)
    eval_loss2.append(val_losss)
    performance_matrix(image_targets_all, image_preds_all)

    return epoch_loss / len(iterator), epoch_acc / len(iterator), val_accu2, eval_loss2
```

```python
EPOCHS_FUSION = 20

train_acc_gr2 = []
train_loss_gr2 = []
val_acc_gr2 = []
val_loss_gr2 = []

for epoch in trange(EPOCHS_FUSION, desc="Fusion Epochs"):
    start_time = time.monotonic()

    train_loss, train_acc, train_acc_gr2, train_loss_gr2 = train_fusion(
        model2, train_radio_loader, optimizer2, criterion2, device)
    valid_loss, valid_acc, val_acc_gr2, val_loss_gr2 = evaluate_fusion(
        model2, val_radio_loader, criterion2, device)

    if epoch % 30 == 0:
        checkpoint_model(epoch, model2, optimizer2, 0., '/kaggle/working/fusion_epoch%d.pth' % epoch)

    end_time = time.monotonic()
    epoch_mins, epoch_secs = epoch_time(start_time, end_time)

    print(f'Epoch: {epoch+1:02} | Epoch Time: {epoch_mins}m {epoch_secs}s')
    print(f'\tTrain Loss: {train_loss:.3f} | Train Acc: {train_acc*100:.2f}%')
    print(f'\t Val. Loss: {valid_loss:.3f} |  Val. Acc: {valid_acc*100:.2f}%')
```

```python
class GradCAM:
    """Grad-CAM for the fusion model. Hooks into the last conv feature maps
    produced by forward_cnn (after denseblock4 + norm5 + relu)."""

    def __init__(self, model):
        self.model = model
        self.gradients = None
        self.activations = None

    def _save_gradient(self, grad):
        self.gradients = grad

    def __call__(self, img_tensor, radiomics_tensor, target_class=None):
        self.model.eval()

        # Forward through CNN to get feature maps
        feat_maps = self.model.forward_cnn(img_tensor)
        feat_maps.register_hook(self._save_gradient)
        self.activations = feat_maps

        # Global pool + fusion + classifier
        cnn_feats = F.adaptive_avg_pool2d(feat_maps, 1).flatten(1)
        radio_feats = self.model.radiomics_branch(radiomics_tensor)
        fused = torch.cat([cnn_feats, radio_feats], dim=1)
        logits = self.model.classifier(fused)

        if target_class is None:
            target_class = logits.argmax(dim=1).item()

        # Backward
        self.model.zero_grad()
        one_hot = torch.zeros_like(logits)
        one_hot[0, target_class] = 1.0
        logits.backward(gradient=one_hot, retain_graph=True)

        # Weight feature maps by mean gradient per channel
        weights = self.gradients.mean(dim=(2, 3), keepdim=True)
        cam = (weights * self.activations).sum(dim=1, keepdim=True)
        cam = F.relu(cam)
        cam = F.interpolate(cam, size=img_tensor.shape[2:], mode='bilinear', align_corners=False)
        cam = cam.squeeze().detach().cpu().numpy()

        # Normalize to [0, 1]
        cam = cam - cam.min()
        if cam.max() > 0:
            cam = cam / cam.max()

        return cam, target_class


def show_gradcam(img_path, model, transform, labels_dict, device):
    """Display Grad-CAM heatmap overlaid on the original image."""
    pil_img = Image.open(img_path).convert('RGB')
    img_tensor = transform(pil_img).unsqueeze(0).to(device)

    radiomics = extract_radiomics(pil_img)
    radio_tensor = torch.tensor(radiomics, dtype=torch.float32).unsqueeze(0).to(device)

    gradcam = GradCAM(model)
    cam, pred_class = gradcam(img_tensor, radio_tensor)

    img_np = np.array(pil_img.resize((256, 256))) / 255.0
    heatmap = plt.cm.jet(cam)[:, :, :3]
    overlay = 0.5 * img_np + 0.5 * heatmap

    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    axes[0].imshow(img_np)
    axes[0].set_title('Original Image')
    axes[0].axis('off')

    axes[1].imshow(cam, cmap='jet')
    axes[1].set_title('Grad-CAM Heatmap')
    axes[1].axis('off')

    axes[2].imshow(overlay)
    axes[2].set_title(f'Overlay - Predicted: {labels_dict[pred_class]}')
    axes[2].axis('off')

    plt.tight_layout()
    plt.show()
    print(f"Predicted class: {labels_dict[pred_class]}")
```
