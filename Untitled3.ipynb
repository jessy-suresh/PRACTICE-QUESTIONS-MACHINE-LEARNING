```python
import kagglehub
import pandas as pd
import os

path = kagglehub.dataset_download("imtkaggleteam/netflix")
print("Dataset path:", path)

csv_file = [f for f in os.listdir(path) if f.endswith(".csv")][0]
df = pd.read_csv(os.path.join(path, csv_file))

print(df.columns)
df.head()
#import the dataset into colab
```

```python
docs = df["description"].dropna().astype(str).tolist()
print("Total descriptions:", len(docs))
```

```python
selected_docs = [d for d in docs if 200 <= len(d.split()) <= 500][:5]

print("Selected docs:", len(selected_docs))
print(selected_docs[0][:400])
```

```python
combined_docs = []
current_doc = ""

for desc in docs:
    if len(current_doc.split()) < 200:
        current_doc += " " + desc
    else:
        if 200 <= len(current_doc.split()) <= 500:
            combined_docs.append(current_doc.strip())
        current_doc = desc

# Ensure we only take 5
selected_docs = combined_docs[:5]

print("Selected docs:", len(selected_docs))
print("Word count of first doc:", len(selected_docs[0].split()))
print(selected_docs[0][:500])
```

```python
import re

def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-z\s]', ' ', text)
    return re.sub(r'\s+', ' ', text).strip()

cleaned_docs = [clean_text(doc) for doc in selected_docs]

print("CLEANED SAMPLE:\n")
print(cleaned_docs[0][:500])
```

```python
import nltk
nltk.download('punkt')

token_lists = [nltk.word_tokenize(doc) for doc in cleaned_docs]

print("TOKENS SAMPLE:\n")
print(token_lists[0][:25])
```

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')

```

```python
token_lists = [nltk.word_tokenize(doc) for doc in cleaned_docs]
print(token_lists[0][:25])
```

```python
from nltk.stem import PorterStemmer, WordNetLemmatizer
nltk.download('wordnet')

stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

sample_words = token_lists[0][:15]

comparison = [(w, stemmer.stem(w), lemmatizer.lemmatize(w)) for w in sample_words]

import pandas as pd
pd.DataFrame(comparison, columns=["Original", "Stemmed", "Lemmatized"])
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(stop_words='english')
tf_matrix = cv.fit_transform(cleaned_docs)

tf_df = pd.DataFrame(tf_matrix.toarray(), columns=cv.get_feature_names_out())
tf_df.iloc[:, :10]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(cleaned_docs)

idf_values = dict(zip(tfidf.get_feature_names_out(), tfidf.idf_))
idf_df = pd.DataFrame(idf_values.items(), columns=["Word", "IDF"]).sort_values(by="IDF", ascending=False).head(10)

tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf.get_feature_names_out())

idf_df.head(), tfidf_df.iloc[:, :10]
```

```python
bigram_vec = CountVectorizer(ngram_range=(2,2), stop_words='english')
bigram_matrix = bigram_vec.fit_transform(cleaned_docs)

bigram_freq = pd.DataFrame(
    bigram_matrix.sum(axis=0).A1,
    index=bigram_vec.get_feature_names_out(),
    columns=["Frequency"]
).sort_values(by="Frequency", ascending=False).head(10)

bigram_freq
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")

pos_counts = {}

for doc in cleaned_docs:
    for token in nlp(doc):
        pos_counts[token.pos_] = pos_counts.get(token.pos_, 0) + 1

pos_df = pd.DataFrame(pos_counts.items(), columns=["POS", "Count"]).sort_values(by="Count", ascending=False)
pos_df
```

```python
from collections import Counter

all_tokens = [word for doc in token_lists for word in doc]
counts = Counter(all_tokens)
total = sum(counts.values())
vocab = len(counts)

prob_data = []
for word in list(counts.keys())[:10]:
    unsmoothed = counts[word] / total
    smoothed = (counts[word] + 1) / (total + vocab)
    prob_data.append((word, unsmoothed, smoothed))

prob_df = pd.DataFrame(prob_data, columns=["Word", "Unsmoothed P", "Smoothed P"])
prob_df
```
